# 🔍 Analyze Submission File

Before uploading predictions to Kaggle,
we verify:
- Data format
- Prediction quality
- Missing values
- Infinite values
- Statistical distribution

This helps detect:
- Model issues
- Overflow problems
- Invalid predictions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Submission file

```python
submission = pd.read_csv("submission.csv")
submission.head()
```

#  Statistical Summary

We analyze:
- Minimum prediction
- Maximum prediction
- Average predicted price
- Distribution spread

```python
# =========================================================
# SUBMISSION STATISTICS
# =========================================================

submission["SalePrice"].describe()
```

#  Check for Invalid Predictions

We now verify:
- Missing values
- Infinite values
- Negative prices

Invalid predictions can cause:
- Kaggle submission errors
- Poor leaderboard score

```python
# =========================================================
# CHECK MISSING VALUES
# =========================================================

submission.isnull().sum()
```

```python
# =========================================================
# CHECK INFINITE VALUES
# =========================================================

np.isinf(
    submission["SalePrice"]
).sum()
```

```python
# =========================================================
# CHECK NEGATIVE VALUES
# =========================================================

(
    submission["SalePrice"] < 0
).sum()
```

#  Prediction Distribution Visualization

We visualize:
- Prediction spread
- Price distribution
- Outliers
- Skewness

```python
# =========================================================
# PREDICTION DISTRIBUTION
# =========================================================

plt.figure(figsize=(10,5))

sns.histplot(
    submission["SalePrice"],
    kde=True
)

plt.title("Predicted SalePrice Distribution")

plt.show()
```

#  Boxplot Analysis

Boxplots help identify:
- Extreme outliers
- Unusual predictions
- Prediction stability

```python
# =========================================================
# BOXPLOT
# =========================================================

plt.figure(figsize=(10,4))

sns.boxplot(
    x=submission["SalePrice"]
)

plt.title("SalePrice Prediction Boxplot")

plt.show()
```
